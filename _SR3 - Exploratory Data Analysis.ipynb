```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np


# Load the dataset
file_path =( r"C:\Users\pac computers\Downloads\InformalDataset.csv.csv")
df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Location')
plt.title('Number of Settlements per Location')
plt.xlabel('Location')
plt.ylabel('Number of Settlements')
plt.xticks(rotation=45)
plt.show()
```

```python
sanitation_counts = df['Sanitation'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sanitation_counts, labels=sanitation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sanitation Availability')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Initial Population'], bins=20, kde=True)
plt.title('Distribution of Initial Population')
plt.xlabel('Initial Population')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x='Location', y='Current Population')
plt.title('Initial Population by Location')
plt.xlabel('Location')
plt.ylabel('Current Population')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Initial Population', y='Current Population', hue='Location')
plt.title('Initial Population vs Current Population')
plt.xlabel('Initial Population')
plt.ylabel('Current Population')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python

# Convert 'Current Population' to numeric, forcing errors to NaN
df['Current Population'] = pd.to_numeric(df['Current Population'], errors='coerce')

# Display summary statistics for 'Initial Population' and 'Current Population'
summary_statistics = df[['Initial Population', 'Current Population']].describe()
print(summary_statistics)



```
