```python
# Check for missing values by checking each column manually
print("\nChecking for missing values manually:")

# Define a function to check for missing values manually
def check_missing_values(dataframe):
    for column in dataframe.columns:
        missing_count = dataframe[column].isnull().sum()
        if missing_count > 0:
            print(f"Column '{column}' has {missing_count} missing values.")

# Use the function to check for missing values
check_missing_values(df)
```

```python
def calculate_total_population(data):
    total_population = 0
    for settlement in data:
        current_population = settlement['Current Population']
        if current_population != "Unknown":
            total_population += int(current_population.replace(',', ''))
    return total_population
```
